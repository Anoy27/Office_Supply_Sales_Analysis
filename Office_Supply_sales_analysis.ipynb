# Feature Engineering

In this section, I have added a few columns for the analysis. Firstly, I have added a new loss column as the profit column has some negative values so added them inside this loss column. Then I gradually did some calculations and added the Product_Price, Profit_Margin,  Delivery_Time,Year and Discount_Amount columns to find insigts from the dataset

```python
d['Loss'] = d['Profit'].apply(lambda x: x if x < 0 else 0)
d["Loss"]=d["Loss"].abs()
```

```python
d["Product_Price"]=d["Sales"]/(d["Quantity"]*(1-d["Discount"]))
```

```python
d["Profit_Margin"]=(d["Profit"]/d["Sales"])
```

```python
d['Delivery_Time'] = d['Ship_Date'] - d['Order_Date']
```

```python
d["Year"]=d["Ship_Date"].dt.year
```

```python
d["Discount_Amount"]=d["Product_Price"]*d["Discount"]
```
